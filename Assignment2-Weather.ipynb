## 1.Weather Data

Dataset from: NOAA
https://www.ncdc.noaa.gov/cdo-web/datasets/GHCND/stations/GHCND:USW00094728/detail

Data period: 2018-01-01 - 2018-12-31

All data is from station USW00094728 (NY CITY CENTRAL PARK, NY US)
We assume the weather from central park could represent the majority part of our data
Since most of transaction happens in Manhattan.


This Data include 
 

WT01 - Fog, ice fog, or freezing fog (may include heavy fog)
WT02 - Heavy fog or heaving freezing fog (not always distinguished from fog)
WT03 - Thunder
WT06 - Glaze or rime
WT08 - Smoke or haze


TMAX - Maximum temperature
TAVG - Average Temperature.
TMIN - Minimum temperature

AWND - Average wind speed
SNOW - Snowfall
PRCP - Precipitation
SNWD - Snow depth

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fahr_to_celsius(temp_fahr):
    """Convert Fahrenheit to Celsius    
    Return Celsius conversion of input"""
    
    temp_celsius = (temp_fahr - 32) * 5 / 9
    return temp_celsius
```

```python
weather = pd.read_csv("E:/Uni/MAST30034/Data/2018_NewYork_weather.csv")
```

```python
print(weather)
```

```python
data = weather.drop(["STATION","NAME", "LATITUDE", "LONGITUDE", "ELEVATION", "AWND_ATTRIBUTES",
                     "SNOW_ATTRIBUTES","SNWD_ATTRIBUTES","TAVG_ATTRIBUTES","TMAX_ATTRIBUTES",
                     "PRCP_ATTRIBUTES", "WT01_ATTRIBUTES", "WT02_ATTRIBUTES", "WT03_ATTRIBUTES",
                    "WT06_ATTRIBUTES","WT08_ATTRIBUTES" ,"TMIN_ATTRIBUTES"], axis='columns')
data
```

From our observation we found that the temperture are recorded in Fahrenheit,
We convert Fahrenheit to Celsius and calculate the average temperture.
Also, we label the weather type for each day,
we define :
Fog, ice fog, or freezing fog (may include heavy fog)
Heavy fog or heaving freezing fog (not always distinguished from fog)
Thunder
Glaze or rime
Smoke or haze
as bad weathers

```python
data["TMAX"] = fahr_to_celsius(data["TMAX"])
data["TMIN"] = fahr_to_celsius(data["TMIN"])
data["TAVG"] = (data["TMAX"] +data["TMIN"])/2
data["Weather Type"] = data[["WT01","WT01","WT03","WT06","WT08"]].sum(axis=1)
data = data.drop(["WT01", "WT02", "WT03", "WT06", "WT08","TMAX","TMIN"], axis='columns')
data['Weather Type'] = np.where(data['Weather Type']>0, 'Y', 'N')

data = data.fillna(0)
```
